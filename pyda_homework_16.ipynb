Написать **линейную регрессию** на два атрибута. Допустим, мы хотим предсказать цену дома (MEDV) в Бостоне, зная что цена на дом имеет линейную зависимость с атрибутом LSTAT. 



1.   Разделить данные на 80% тренировочных, 20% тестовых
2.   Составить модель линейной регрессии
3.   Посчитать ошибку модели на тестовых данных
4.   Сделать точечный график данных и полученной прямой линейной регрессии



```python
MEDV = data['MEDV']
LSTAT = data['LSTAT']
MEDV = np.array(MEDV).reshape(-1,1)
LSTAT = np.array(LSTAT).reshape(-1,1)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
from sklearn.model_selection import train_test_split

# x = MEDV,
# y = LSTAT
X_train, X_test, y_train, y_test = train_test_split(
    MEDV,
    LSTAT,
    test_size = 0.2
)
```

```python
model.fit(X_train, y_train)
predicted_y = model.predict(X_test)
model.score(X_test, y_test)
```

```python
plt.xlabel('MEDV')
plt.ylabel('LSTAT')
plt.title('Линейная регрессия от одной переменной')
# Построим прямую для регрессии
# Создадим 100 точек для предсказаний
x = np.linspace(-50, 200, 100)
# Посчитаем предсказанные значения
# y = kx + b
# k = model.coef_
# b = model.intercept_
y = model.coef_*x + model.intercept_
y = np.transpose(y)
plt.plot(x,y)
plt.plot(X_train, y_train, 'r.')
plt.plot(X_test, y_test, 'go')
plt.xlim(0,55)
plt.ylim(0,40)
plt.legend(['Прямая линейной регрессии', 'Тренировочные данные', 'Тестовые данные'])
plt.show()
```

```python
import datapackage
import pandas as pd
```

```python
#воспользовался сторонним датасетом, уже содержащим разделение стран на континенты
data_url = 'https://datahub.io/JohnSnowLabs/country-and-continent-codes-list/datapackage.json'

# to load Data Package into storage
package = datapackage.Package(data_url)

# to load only tabular data
resources = package.resources
for resource in resources:
    if resource.tabular:
        continent_data = pd.read_csv(resource.descriptor['path'])
    
```

```python
# страны в этом датасете выписаны в полных названиях вида "Iran, Islamic republic of", 
# поэтому разделяю по запятой и скобке, чтобы выбрать общеупотребимое название страны
# Также нужно исправить разногласия у стран, в которых неоднозначна принадлежность к континенту, 
# как у Азербайджана - Азия/Европа, в основном это касается стран у которых неоднозначность Азия/Европа
# Удалим дубликаты, выбирая Азию 
continent_data = continent_data[ ['Continent_Name','Country_Name'] ]
continent_data['Country_Name'] = continent_data['Country_Name'].apply(lambda word: word.split(',')[0]).apply(lambda word: word.split('(')[0])
continent_data.drop_duplicates(subset = 'Country_Name', keep = 'last', inplace = True)
continent_data.reset_index()[['Continent_Name', 'Country_Name']].head()
```

# Дискретное распределение случайной величины

Напишите код для моделирования 10 подбрасываний игрального кубика. Начертите диаграму histogram. Какое это распределение?

```python
import random
def cubic_throwing(throw_number):
    results = []
    for throw in range(throw_number):
        result = random.randint(1,6)
        results.append(result)
    res = pd.DataFrame(
        {
            'result' : results
        }
    )
    return res
cubic_throwing(10)
```

```python
n, bins, patches = plt.hist(cubic_throwing(10)['result'], 6, facecolor='g')

plt.xlabel('frequency')
plt.ylabel('result')
plt.title('Частотный график result')
plt.grid(True)
plt.show()
```

Похоже на равномерное распределение

моделируйте 1000 раз сумму по 100 подбрасываний. Начертите диаграму histogram. Почему график изменился? Почему можно применить центральную предельную теорему?

```python
#второй вариант решения первой задачи
number_of_randoms = 10
plt.hist(np.random.random_integers(1, 6, number_of_randoms))
plt.show()


```

```python
number_of_randoms = 100
number_of_sums = 1000
X1 = np.zeros([number_of_sums])
for i in range(number_of_sums):
    # суммируем все элементы равномерных выборок
    X1[i] = np.sum(np.random.random_integers(1, 6, number_of_randoms))#/number_of_randoms
plt.hist(X1, 50)
plt.xlabel('Сумма результатов 100 бросков')
plt.ylabel('Частота появления')
plt.show()
```

График изменился и стал похож на нормальное распределение, поскольку каждая выборка из 100 бросков не зависит от другой, а результат самого броска является случайной величиной, поэтому и работает закон больших чисел.
